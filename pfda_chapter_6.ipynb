```python
# Manually working with delimited formats
!cat ch06/ex7.csv
```

```python
import csv
f = open('ch06/ex7.csv')
reader = csv.reader(f)
```

```python
for line in reader:
    print line
```

```python
lines = list(csv.reader(open('ch06/ex7.csv')))
header, values = lines[0], lines[1:]
data_dict = {h: v for h, v in zip(header, zip(*values))}
data_dict
```

```python
# Defining a new CSV format
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL
reader = csv.reader(f, dialect = my_dialect)
```

```python
reader = csv.reader(f, delimiter = '|')
```

```python
# Writing delimited files manual
with open('ch06/mydata.csv', 'w') as f:
    writer = csv.writer(f, dialect = my_dialect)
    writer.writerow(('one', 'two', 'three'))
    writer.writerow(('1', '2', '3'))
    writer.writerow(('4', '5', '6'))
    writer.writerow(('7', '8', '9'))
```

```python
# XML and HTML: web scraping
# Using the example from http://docs.python-guide.org/en/latest/scenarios/scrape/
# as the link from the book has changed and no longer works
from lxml import html
import requests
```

```python
page = requests.get('http://econpy.pythonanywhere.com/ex/001.html')
tree = html.fromstring(page.content)
```

If youâ€™re using Chrome, you can right click an element, choose â€˜Inspect elementâ€™, highlight the code, right click again and choose â€˜Copy XPathâ€™.

```python
# E.g., 
# <div title="buyer-name">Carson Busses</div>
# <span class="item-price">$29.95</span>
```

```python
# We know there are buyers and prices from inspecting the html element
# This will create a list of buyers:
buyers = tree.xpath('//div[@title="buyer-name"]/text()')
# This will create a list of prices
prices = tree.xpath('//span[@class="item-price"]/text()')
```

```python
print 'Buyers: ', buyers
```

```python
print 'Prices: ', prices
```
